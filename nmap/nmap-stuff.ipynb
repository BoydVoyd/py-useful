```python
from libnmap.parser import NmapParser
```

```python
nmap_report = NmapParser.parse_fromfile('dns_other_up.xml')
```

```python
for host in nmap_report.hosts:
    if host.hostnames:
        s = ', ' + host.hostnames[0]
    else:
        s = ''
    print host.address + s
    for service in host.services:
        print '  ' + str(service)
```

```python
with open('dns_other_report.txt', 'w') as f:
    f.write(s)
```
